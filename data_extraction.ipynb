```python
# imports
import pandas as pd
import os
import shutil
import random as rd
```

```python
# define the path where the raw data is stored
file_path_cp_raw = "raw_data/covid_positive_raw_data/images"
file_path_cn_raw = "raw_data/covid_negative_raw_data/chest_xray/train/normal"
```

```python
# show the csv file for the positve samples
df_pos = pd.read_csv("raw_data/covid_positive_raw_data/metadata.csv")
df_pos.head()
```

```python
# extract images for positive corona x-rays
#
# create path to copy image to
copy_path_covid_positive = "model_data/data_train/covid_positive"
# delete the folder if it already exists to remove all its content
if os.path.exists(copy_path_covid_positive):
    shutil.rmtree(copy_path_covid_positive)
if not os.path.exists(copy_path_covid_positive):
    os.makedirs(copy_path_covid_positive)
    print("Path "+ copy_path_covid_positive+ " created!")

num_pos_sample=0
for n in range(df_pos.shape[0]):
    if (df_pos["finding"][n]=="Pneumonia/Viral/COVID-19" and df_pos["view"][n]=="PA"):
        image_path = os.path.join(file_path_cp_raw,df_pos["filename"][n]) 
        # copy image file into working folder
        shutil.copy(image_path,copy_path_covid_positive)
        # number of images copied
        num_pos_sample = num_pos_sample+1
print("In total "+ str(num_pos_sample) +" of positive corona x-rays have been copied!")        
```

```python
# extract images for negative corona x-rays
#
# create path to copy image to
copy_path_covid_negative = "model_data/data_train/covid_negative"
# delete the folder if it already exists to remove all its content
if os.path.exists(copy_path_covid_negative):
    shutil.rmtree(copy_path_covid_negative)
if not os.path.exists(copy_path_covid_negative):
    os.makedirs(copy_path_covid_negative)
    print("Path "+ copy_path_covid_negative+ " created!")
    
# list of all negative corona images   
imageList = os.listdir(file_path_cn_raw)
# randomly pick negative x-ray images equal to the number of postive x-ray images
num = len(imageList)
rdList = []
num_neg_sample = 0
while num_neg_sample<num_pos_sample:
    # generate random number in range
    rdNum = rd.randint(0,num-1)
    if rdNum not in rdList:
        rdList.append(rdNum)
        image_path = os.path.join(file_path_cn_raw,imageList[rdNum])
        # copy image file into working folder
        shutil.copy(image_path,copy_path_covid_negative)
        num_neg_sample = num_neg_sample+1
print("In total "+ str(num_neg_sample) +" of negative corona x-rays have been copied!")        
```

```python
# create test data from the filtered datasets
train_data_pos_path = "model_data/data_test/covid_positive"
train_data_neg_path = "model_data/data_test/covid_negative"

# define fraction of overall data to use for training
test_size = 0.3

# delete the folder if it already exists to remove all its content
# pos
if os.path.exists(train_data_pos_path):
    shutil.rmtree(train_data_pos_path)
if not os.path.exists(train_data_pos_path):
    os.makedirs(train_data_pos_path)
    #print("Path "+ train_data_pos_path+ " created!")
    
# neg
if os.path.exists(train_data_neg_path):
    shutil.rmtree(train_data_neg_path)
if not os.path.exists(train_data_neg_path):
    os.makedirs(train_data_neg_path)
    #print("Path "+ train_data_pos_path+ " created!")
    
img_pos = os.listdir(copy_path_covid_positive)
img_neg = os.listdir(copy_path_covid_negative)
num_img_pos = len(img_pos)
rdList = []
num_data = 0
# set the ratio between pos/neg data to 1
while num_data<round(test_size*num_img_pos):
    # generate random number in range
    rdNum = rd.randint(0,num_img_pos-1)
    if rdNum not in rdList:
        rdList.append(rdNum)
        image_path_pos = os.path.join(copy_path_covid_positive,img_pos[rdNum])
        image_path_neg = os.path.join(copy_path_covid_negative,img_neg[rdNum])
        # move image file into test folder
        shutil.move(image_path_pos,train_data_pos_path)
        shutil.move(image_path_neg,train_data_neg_path)
        num_data = num_data+1
        
# copy the additional COVID-negative test xrays from the raw data folder into the model data folder for testing
path = "raw_data/covid_negative_raw_data/chest_xray/test/NORMAL"
# number of images to copy
numImg = 40
imgList = os.listdir(path)
for n in range(0,numImg):
    rawPath = os.path.join(path,imgList[n])
    shutil.copy(rawPath,train_data_neg_path)

print("Training data has been generated!")
```
